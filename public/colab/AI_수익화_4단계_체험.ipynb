# üé® Step 2: Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± (ÎÇòÎÖ∏ Î∞îÎÇòÎÇò üçå)

GoogleÏùò **Gemini 2.0 Flash**Î°ú Ïù¥ÎØ∏ÏßÄÎ•º ÏÉùÏÑ±Ìï©ÎãàÎã§!

Ïù¥Í≤å Î∞îÎ°ú **ÎÇòÎÖ∏ Î∞îÎÇòÎÇò**ÏòàÏöî. Ï∫êÎ¶≠ÌÑ∞ ÏùºÍ¥ÄÏÑ±ÏùÑ Ïú†ÏßÄÌïòÎ©¥ÏÑú Îã§ÏñëÌïú Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì§ Ïàò ÏûàÏñ¥Ïöî.


```python
# üé® Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÏòàÏ†ú
from IPython.display import display
import base64
from io import BytesIO
from PIL import Image as PILImage

# Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± Î™®Îç∏
image_model = genai.GenerativeModel('gemini-2.0-flash-exp-image-generation')

# üéØ Ïó¨Í∏∞ ÌîÑÎ°¨ÌîÑÌä∏Î•º Î∞îÍøîÎ≥¥ÏÑ∏Ïöî!
image_prompt = """
ÌïúÍµ≠Ïùò Ï†ÑÌÜµ ÌïúÏò• ÎßàÏùÑÏóêÏÑú ÌïúÎ≥µÏùÑ ÏûÖÏùÄ Í∑ÄÏó¨Ïö¥ ÏÉ¥ Í≥†ÏñëÏù¥Î•º ÎßåÎì§Ïñ¥ Ï£ºÏÑ∏Ïöî.
Ï¥àÎ°ùÏÉâ ÏôºÏ™Ω ÎààÍ≥º ÌååÎûÄÏÉâ Ïò§Î•∏Ï™Ω ÎààÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÏäµÎãàÎã§.
ÎîîÏ¶àÎãà ÌîΩÏÇ¨ Ïä§ÌÉÄÏùºÎ°ú Í∑∏Î†§Ï£ºÏÑ∏Ïöî.
"""

response = image_model.generate_content(
    image_prompt,
    generation_config={
        "response_modalities": ["image", "text"]
    }
)

# Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
for part in response.candidates[0].content.parts:
    if hasattr(part, 'inline_data') and part.inline_data:
        image_data = base64.b64decode(part.inline_data.data)
        img = PILImage.open(BytesIO(image_data))
        display(img)
        print("\n‚úÖ Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± ÏôÑÎ£å! Ïò§Î•∏Ï™Ω ÌÅ¥Î¶≠Ìï¥ÏÑú Ï†ÄÏû•ÌïòÏÑ∏Ïöî~")
```

### üéØ Ïù¥ÎØ∏ÏßÄ Ìé∏ÏßëÌï¥Î≥¥Í∏∞

ÎÇòÎÖ∏ Î∞îÎÇòÎÇòÏùò ÏßÑÏßú Ìûò! **Ï∫êÎ¶≠ÌÑ∞Î•º Ïú†ÏßÄÌïòÎ©¥ÏÑú Ïû•Î©¥ÏùÑ Î∞îÍøÄ Ïàò ÏûàÏñ¥Ïöî.**


```python
# üîÑ Ïù¥ÎØ∏ÏßÄ Ìé∏Ïßë (Ï∫êÎ¶≠ÌÑ∞ Ïú†ÏßÄÌïòÎ©¥ÏÑú Ïû•Î©¥ Î≥ÄÍ≤Ω)

# Ï±ÑÌåÖ Î™®ÎìúÎ°ú Ïù¥ÎØ∏ÏßÄ Ìé∏Ïßë
chat = image_model.start_chat(history=[])

# Ï≤´ Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±
first_response = chat.send_message(
    "Í∑ÄÏó¨Ïö¥ ÌîåÎùºÏä§Ìã± Ïó¨Ïö∞ ÌîºÍ∑úÏñ¥Î•º ÎßåÎì§Ïñ¥Ï£ºÏÑ∏Ïöî. ÏïÑÏù¥ Î∞©Ïóê ÏûàÎäî Ïû•ÎÇúÍ∞ê Ïä§ÌÉÄÏùºÎ°úÏöî.",
    generation_config={"response_modalities": ["image", "text"]}
)

print("ü¶ä Ï≤´ Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄ:")
for part in first_response.candidates[0].content.parts:
    if hasattr(part, 'inline_data') and part.inline_data:
        image_data = base64.b64decode(part.inline_data.data)
        img = PILImage.open(BytesIO(image_data))
        display(img)
```

```python
# üîÑ Í∞ôÏùÄ Ï∫êÎ¶≠ÌÑ∞Î°ú Ïû•Î©¥ Î≥ÄÍ≤Ω!

edit_response = chat.send_message(
    "Ïù¥ Ïó¨Ïö∞Í∞Ä ÏäàÌçºÌûàÏñ¥Î°ú ÎßùÌÜ†Î•º ÏûÖÍ≥† ÌïòÎäòÏùÑ ÎÇ†Í≥† ÏûàÎäî Î™®ÏäµÏúºÎ°ú Î∞îÍøîÏ£ºÏÑ∏Ïöî!",
    generation_config={"response_modalities": ["image", "text"]}
)

print("ü¶∏ Îëê Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄ (Í∞ôÏùÄ Ï∫êÎ¶≠ÌÑ∞, Îã§Î•∏ Ïû•Î©¥):")
for part in edit_response.candidates[0].content.parts:
    if hasattr(part, 'inline_data') and part.inline_data:
        image_data = base64.b64decode(part.inline_data.data)
        img = PILImage.open(BytesIO(image_data))
        display(img)

print("\n‚ú® Ï∫êÎ¶≠ÌÑ∞Îäî Í∑∏ÎåÄÎ°úÏù∏Îç∞ Ïû•Î©¥Îßå Î∞îÎÄåÏóàÏ£†? Ïù¥Í≤å ÎÇòÎÖ∏ Î∞îÎÇòÎÇòÏùò ÌûòÏûÖÎãàÎã§!")
```

Ïã§ÏãúÍ∞Ñ Ï†ïÎ≥¥Î•º Í≤ÄÏÉâÌï¥ÏÑú Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì§Ïñ¥Ï§òÏöî. ÏòàÎ•º Îì§Ïñ¥:

```python
# üå¶Ô∏è Ïã§ÏãúÍ∞Ñ ÎÇ†Ïî® ÏòàÎ≥¥ Ï∞®Ìä∏ ÏÉùÏÑ±!
from google import genai
from google.genai import types

# ÏÉàÎ°úÏö¥ GenAI ÌÅ¥ÎùºÏù¥Ïñ∏Ìä∏ (Pro Î™®Îç∏Ïö©)
client = genai.Client(api_key=GOOGLE_API_KEY)

# üéØ ÎèÑÏãú Ïù¥Î¶ÑÏùÑ Î∞îÍøîÎ≥¥ÏÑ∏Ïöî!
city = "ÏÑúÏö∏"

prompt = f"{city}Ïùò Ìñ•ÌõÑ 5ÏùºÍ∞Ñ ÎÇ†Ïî® ÏòàÎ≥¥Î•º ÍπîÎÅîÌïú Ï∞®Ìä∏Î°ú ÏãúÍ∞ÅÌôîÌï¥Ï£ºÏÑ∏Ïöî. Í∞Å ÎÇ†ÏßúÎ≥ÑÎ°ú Î≠ò ÏûÖÏúºÎ©¥ Ï¢ãÏùÑÏßÄÎèÑ Î≥¥Ïó¨Ï£ºÏÑ∏Ïöî!"

response = client.models.generate_content(
    model="gemini-3-pro-image-preview",
    contents=prompt,
    config=types.GenerateContentConfig(
        response_modalities=['Text', 'Image'],
        image_config=types.ImageConfig(aspect_ratio="16:9"),
        tools=[{"google_search": {}}]  # üî• Google Í≤ÄÏÉâ ÌôúÏÑ±Ìôî!
    )
)

print(f"üå¶Ô∏è {city} ÎÇ†Ïî® ÏòàÎ≥¥ Ï∞®Ìä∏:\n")
for part in response.candidates[0].content.parts:
    if hasattr(part, 'text') and part.text:
        print(part.text)
    elif hasattr(part, 'inline_data') and part.inline_data:
        image_data = base64.b64decode(part.inline_data.data)
        img = PILImage.open(BytesIO(image_data))
        display(img)

print("\n‚ú® Ïã§ÏãúÍ∞Ñ Îç∞Ïù¥ÌÑ∞Î•º Í≤ÄÏÉâÌï¥ÏÑú Ïù¥ÎØ∏ÏßÄÎ•º ÎßåÎì§ÏóàÏñ¥Ïöî! Google Search GroundingÏùò Ìûò!")
```
